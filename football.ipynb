```python
import pandas as pd
import matplotlib.pyplot as plt

# قراءة البيانات من ملف CSV
data = pd.read_csv(r'C:\Users\SaMy\Desktop\Data.csv')  # تأكد من أن المسار صحيح

# عرض أول 5 صفوف من البيانات
print(data.head())

# 1. تحليل أفضل الهدافين
# تأكد من أن أسماء الأعمدة صحيحة، هنا افترضت أنها 'Player Names', 'Goals', 'Matches_Played'
top_scorers = data[['Player Names', 'Goals', 'Matches_Played']].copy()
top_scorers['Goals_per_Match'] = top_scorers['Goals'] / top_scorers['Matches_Played']
top_scorers = top_scorers.sort_values(by='Goals', ascending=False).head(10)

print("أفضل الهدافين:")
print(top_scorers)

# 2. مقارنة xG مع الأهداف
plt.figure(figsize=(10, 6))
plt.scatter(data['xG'], data['Goals'], alpha=0.6)
plt.title('مقارنة xG مع الأهداف')
plt.xlabel('xG')
plt.ylabel('Goals')
plt.axline((0, 0), slope=1, color='red', linestyle='--')  # خط يمثل الأداء المتوقع
plt.grid()
plt.show()

# 3. رسم بياني لأداء الهدافين
plt.figure(figsize=(10, 6))
plt.bar(top_scorers['Player Names'], top_scorers['Goals'], color='skyblue')
plt.title('أفضل الهدافين في الدوري')
plt.xlabel('اسم اللاعب')
plt.ylabel('عدد الأهداف')
plt.xticks(rotation=45)
plt.show()
```
